```python
import pandas as pd
import bt as bt
import yfinance as yf
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def consulta_bc(codigo_bcb):
    url = 'http://api.bcb.gov.br/dados/serie/bcdata.sgs.{}/dados?formato=json'.format(codigo_bcb)
    df = pd.read_json(url)
    df['data'] = pd.to_datetime(df['data'], dayfirst=True)
    df.set_index('data', inplace=True)
    return df
```

```python
def cdi_acumulado(data_inicio, data_fim):
    cdi = consulta_bc(12)
    cdi_acumulado = (1 + cdi[data_inicio : data_fim] / 100).cumprod()
    cdi_acumulado.iloc[0] = 1
    return cdi_acumulado
```

```python
data_inicio = '2011-01-01'
data_fim = '2024-10-31'

cdi = cdi_acumulado(data_inicio = data_inicio, data_fim = data_fim)
```

```python
tickers_carteira = ["PETR4.SA", "VALE3.SA", "ITUB4.SA", "CSMG3.SA", "BRKM5.SA", "CPLE6.SA"]

carteira = yf.download(tickers_carteira, start=data_inicio, end=data_fim)['Adj Close']
```

```python
#carteira['renda_fixa'] = cdi -> plotar e verificar onde está o erro aqui
#carteira.dropna(axis='columns') -> remove as colunas com erro
#linhas com erro

carteira.dropna(inplace=True)

df = pd.DataFrame(carteira)
```

```python
carteira
```

```python
rebalanceamento = bt.Strategy('rebalanceamento', 
                            [bt.algos.RunMonthly(run_on_end_of_period=True),
                             bt.algos.SelectAll(),
                             bt.algos.WeighEqually(),
                             bt.algos.Rebalance()])
```

```python
buy_hold = bt.Strategy('Buy&Hold', 
                   [ bt.algos.RunOnce(),
                     bt.algos.SelectAll(),
                     bt.algos.WeighEqually(),
                     bt.algos.Rebalance()]
                    )
```

```python
bt1 = bt.Backtest(rebalanceamento, carteira)
bt2 = bt.Backtest(buy_hold, carteira)
```

```python
resultados = bt.run(bt1, bt2)
```

```python
def plot_security_weights(result, name):
    result.get_security_weights().plot(figsize=(10, 6))
    plt.title(f'{name} Portfolio Weights')
    plt.xlabel('Date')
    plt.ylabel('Weight')
    plt.legend(title='Asset', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.grid()
    plt.show()
```
